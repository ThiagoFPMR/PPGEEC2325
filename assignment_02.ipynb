```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Questão 1

Dado 02 v.a. X e Y Normais, independentes com média zero e variância 1, gere sequências
de 5.000 amostras cada para uma delas. Então:

```python
def plot_scatter(x, y, xlabel="", ylabel="", ax=None):
    """
    Scatter plot using seaborn. If ax is provided, plot on that axis; otherwise, create a new figure.
    """
    import seaborn as sns
    title = f"Scatter plot of {xlabel} vs {ylabel}"
    sns.set_style("whitegrid")
    if ax is None:
        plt.figure(figsize=(8, 6))
        ax = plt.gca()
    sns.scatterplot(x=x, y=y, alpha=0.5, ax=ax)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.axis('equal')
    if ax is None or not plt.fignum_exists(ax.figure.number):
        plt.show()
```

```python
N_SAMPLES = 5000
# Sample X, Y from a normal distribution with mean 0 and std 1
X = np.random.normal(0, 1, N_SAMPLES)
Y = np.random.normal(0, 1, N_SAMPLES)
```

```python
# Subplot all scatter plots in a 5x3 grid
fig, axes = plt.subplots(5, 3, figsize=(15, 25))
axes = axes.flatten()

def add_stats(letter, w_expr, z_expr, corr, cov):
    # Compose a multi-line title with W and Z on separate lines, then stats
    return f"{letter}) W: {w_expr}\nZ: {z_expr}\nCorr={corr:.3f}, Cov={cov:.3f}"

# a) X, Y
corr = np.corrcoef(X, Y)[0, 1]
cov = np.cov(X, Y)[0, 1]
plot_scatter(X, Y, xlabel="X", ylabel="Y", ax=axes[0])
axes[0].set_title(add_stats("a", "X", "Y", corr, cov), wrap=True)

# b) W = 2X + 5, Z = Y
W = 2 * X + 5
Z = Y
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2X + 5)", ylabel="Z (Y)", ax=axes[1])
axes[1].set_title(add_stats("b", "2X + 5", "Y", corr, cov), wrap=True)

# c) W = 2X + 5, Z = Y + 3
W = 2 * X + 5
Z = Y + 3
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2X + 5)", ylabel="Z (Y + 3)", ax=axes[2])
axes[2].set_title(add_stats("c", "2X + 5", "Y + 3", corr, cov), wrap=True)

# d) W = 2(X + 5), Z = Y + 3
W = 2 * (X + 5)
Z = Y + 3
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2*(X + 5))", ylabel="Z (Y + 3)", ax=axes[3])
axes[3].set_title(add_stats("d", "2(X + 5)", "Y + 3", corr, cov), wrap=True)

# e) W = 2cos(q)X - sin(q)Y, Z = 2sin(q)X + cos(q)Y, q=0
q = 0
W = 2 * np.cos(q) * X - np.sin(q) * Y
Z = 2 * np.sin(q) * X + np.cos(q) * Y
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2cos(0)X - sin(0)Y)", ylabel="Z (2sin(0)X + cos(0)Y)", ax=axes[4])
axes[4].set_title(add_stats("e", "2cos(0)X - sin(0)Y", "2sin(0)X + cos(0)Y", corr, cov), wrap=True)

# f) q = pi/4
q = np.pi / 4
W = 2 * np.cos(q) * X - np.sin(q) * Y
Z = 2 * np.sin(q) * X + np.cos(q) * Y
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2cos(π/4)X - sin(π/4)Y)", ylabel="Z (2sin(π/4)X + cos(π/4)Y)", ax=axes[5])
axes[5].set_title(add_stats("f", "2cos(π/4)X - sin(π/4)Y", "2sin(π/4)X + cos(π/4)Y", corr, cov), wrap=True)

# g) q = pi/2
q = np.pi / 2
W = 2 * np.cos(q) * X - np.sin(q) * Y
Z = 2 * np.sin(q) * X + np.cos(q) * Y
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2cos(π/2)X - sin(π/2)Y)", ylabel="Z (2sin(π/2)X + cos(π/2)Y)", ax=axes[6])
axes[6].set_title(add_stats("g", "2cos(π/2)X - sin(π/2)Y", "2sin(π/2)X + cos(π/2)Y", corr, cov), wrap=True)

# h) q = 3pi/4
q = 3 * np.pi / 4
W = 2 * np.cos(q) * X - np.sin(q) * Y
Z = 2 * np.sin(q) * X + np.cos(q) * Y
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2cos(3π/4)X - sin(3π/4)Y)", ylabel="Z (2sin(3π/4)X + cos(3π/4)Y)", ax=axes[7])
axes[7].set_title(add_stats("h", "2cos(3π/4)X - sin(3π/4)Y", "2sin(3π/4)X + cos(3π/4)Y", corr, cov), wrap=True)

# i) W = cos(q)(2X+5) - sin(q)(Y+3), Z = sin(q)(2X+5) + cos(q)(Y+3), q=π/4
q = np.pi / 4
W = (2 * X + 5) * np.cos(q) - (Y + 3) * np.sin(q)
Z = (2 * X + 5) * np.sin(q) + (Y + 3) * np.cos(q)
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (cos(π/4)(2X+5) - sin(π/4)(Y+3))", ylabel="Z (sin(π/4)(2X+5) + cos(π/4)(Y+3))", ax=axes[8])
axes[8].set_title(add_stats("i", "cos(π/4)(2X+5) - sin(π/4)(Y+3)", "sin(π/4)(2X+5) + cos(π/4)(Y+3)", corr, cov), wrap=True)

# j) W = 2cos(q)X - sin(q)Y + 5, Z = 2sin(q)X + cos(q)Y + 3, q=π/4
q = np.pi / 4
W = 2 * np.cos(q) * X - np.sin(q) * Y + 5
Z = 2 * np.sin(q) * X + np.cos(q) * Y + 3
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2cos(π/4)X - sin(π/4)Y + 5)", ylabel="Z (2sin(π/4)X + cos(π/4)Y + 3)", ax=axes[9])
axes[9].set_title(add_stats("j", "2cos(π/4)X - sin(π/4)Y + 5", "2sin(π/4)X + cos(π/4)Y + 3", corr, cov), wrap=True)

# k) W = 2cos(q)X - sin(q)Y + 5, Z = 2sin(q)X + cos(q)Y + 3, q=3π/4
q = 3 * np.pi / 4
W = 2 * np.cos(q) * X - np.sin(q) * Y + 5
Z = 2 * np.sin(q) * X + np.cos(q) * Y + 3
corr = np.corrcoef(W, Z)[0, 1]
cov = np.cov(W, Z)[0, 1]
plot_scatter(W, Z, xlabel="W (2cos(3π/4)X - sin(3π/4)Y + 5)", ylabel="Z (2sin(3π/4)X + cos(3π/4)Y + 3)", ax=axes[10])
axes[10].set_title(add_stats("k", "2cos(3π/4)X - sin(3π/4)Y + 5", "2sin(3π/4)X + cos(3π/4)Y + 3", corr, cov), wrap=True)

# Hide unused axes
for i in range(11, 15):
    axes[i].set_visible(False)

plt.tight_layout()
plt.show()
```

# Questão 2

## Subquestões a, b, e c

```python
N_SAMPLES = 100000

# Define the parameters for bivariate normal distribution
mean = np.array([5, 3])  # Mean vector [μ₁, μ₂]

# Covariance matrix
# [[σ₁², ρσ₁σ₂],
#  [ρσ₁σ₂, σ₂²]]
# where ρ is the correlation coefficient

# Create covariance matrix
cov_matrix = np.array([[4, 1.8],
                       [1.8, 1]])

print("Mean vector:", mean)
print("Covariance matrix:")
print(cov_matrix)

# Generate samples from the bivariate normal distribution
samples = np.random.multivariate_normal(mean, cov_matrix, N_SAMPLES)

print(f"\nGenerated {N_SAMPLES} samples")
print("Sample statistics:")
print("Sample mean:", np.mean(samples, axis=0))
print("Sample covariance matrix:")
print(np.cov(samples.T))        
```

```python
# Create a grid for plotting the PDF

# mins and max from samples
x_min, y_min = samples.min(axis=0)
x_max, y_max = samples.max(axis=0)

x = np.linspace(x_min, x_max, 100)
y = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x, y)

# Create the bivariate normal distribution object
from scipy.stats import multivariate_normal
rv = multivariate_normal(mean, cov_matrix)

# Calculate the PDF values on the grid
pos = np.dstack((X, Y))
pdf_values = rv.pdf(pos)

# Create subplots
fig = plt.figure(figsize=(16, 5))

# 1. Scatter plot of samples with contours
ax1 = fig.add_subplot(131)
ax1.scatter(samples[:5000, 0], samples[:5000, 1], alpha=0.5, s=1)
contours = ax1.contour(X, Y, pdf_values, colors='red', linewidths=2)
ax1.clabel(contours, inline=True, fontsize=8)
ax1.set_xlabel('X₁')
ax1.set_ylabel('X₂')
ax1.set_title('Samples with PDF Contours')
ax1.grid(True, alpha=0.3)

# 2. Filled contour plot
ax2 = fig.add_subplot(132)
contourf = ax2.contourf(X, Y, pdf_values, levels=20, cmap='viridis')
plt.colorbar(contourf, ax=ax2, label='PDF Value')
ax2.set_xlabel('X₁')
ax2.set_ylabel('X₂')
ax2.set_title('Bivariate PDF - Contour Plot')

# 3. 3D surface plot
ax3 = fig.add_subplot(133, projection='3d')
surface = ax3.plot_surface(X, Y, pdf_values, cmap='viridis', alpha=0.8)
ax3.set_xlabel('X₁')
ax3.set_ylabel('X₂')
ax3.set_zlabel('PDF Value')
ax3.set_title('Bivariate PDF - 3D Surface')
plt.colorbar(surface, ax=ax3, shrink=0.8)

plt.tight_layout()
plt.show()
```

```python
q = 3 * np.pi / 4
# Rotation matrix of angle q
R = np.array([[np.cos(q), -np.sin(q)],
                [np.sin(q),  np.cos(q)]])

cov_matrix = R @ cov_matrix @ R.T
print("Rotated covariance matrix:")
print(cov_matrix)

samples = np.random.multivariate_normal(mean, cov_matrix, N_SAMPLES)

print(f"\nGenerated {N_SAMPLES} samples")
print("Sample statistics:")
print("Sample mean:", np.mean(samples, axis=0))
print("Sample covariance matrix:")
print(np.cov(samples.T))      
```
